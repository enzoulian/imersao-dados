````python
import numpy as np
# Compute the correlation matrix
corr = dados.loc[:,'g-0':'g-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
````

````python
# Compute the correlation matrix
corr_celulas = dados.loc[:,'c-0':'c-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_celulas, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
````

# Aula 4

Resultado dos experimentos

````python
dados_resultados = pd.read_csv("https://github.com/alura-cursos/imersao-dados-desafio-final/blob/main/Dados/dados_resultados.csv?raw=true")
dados_resultados.head()
````

````python
dados_resultados['acetylcholine_receptor_agonist'].unique()
````

Contagem dos mecanismos de ação (MOA) pelas colunas

````python
#contagem_moa = dados_resultados.select_dtypes('int64').sum().sort_values(ascending=False) - seleciona e soma todos os valores das colunas (sum axis=0) que são do tipo int
contagem_moa = dados_resultados.drop('id', axis=1).sum(axis=0).sort_values(ascending=False) # - seleciona e soma todos os valores das colunas (sum axis=0), exceto coluna id (id, axis=1)
contagem_moa
````

Contagem dos mecanismos de ação pelas linhas (experimentos)

````python
dados_resultados.drop('id', axis=1).sum(axis=1)
````

Juntando as bases de dados de Experimento e Resultados

````python
dados_resultados['n_moa'] = dados_resultados.drop('id', axis=1).sum(axis=1) #inserir coluna com a soma dos mecanismos ativos por linha
````

````python
dados_resultados['ativo_moa'] = (dados_resultados['n_moa'] !=0 ) #inserir coluna para verificar se foi ativo ou não/ Se ativo, resultado True, se inibido, resultado False
dados_resultados.head()
````

````python
dados_combinados = pd.merge(dados, dados_resultados[['id','n_moa', 'ativo_moa']], on='id') # on='id' está associando as chaves primárias das duas base de dados
dados_combinados.head()
````

````python
dados_combinados.query('tratamento == "com_controle"')['ativo_moa'].value_counts() #buscando os testes com controle que tiveram algum moa ativo
````

````python
dados_combinados.query('tratamento == "com_droga"')['ativo_moa'].value_counts() #buscando os testes com droga que tiveram algum moa ativo
````

````python
composto_principal = dados_combinados['composto'].value_counts().index[:5] #pegando os 5 compostos mais utilizados
plt.figure(figsize=(12,8))
sns.boxplot(data = dados_combinados.query('composto in @composto_principal'), y='g-0', x='composto', hue='ativo_moa')
````
